```python
movie["actor_likes"]=movie[['actor_2_facebook_likes','actor_1_facebook_likes','actor_3_facebook_likes']].sum(axis=1)
```

### 2. Use crosstabs to show the number of moovies per country

```python
movie.reset_index(inplace=True)
```

```python
pd.crosstab(movie.country,movie.color,values=movie.index, aggfunc='count',margins=True,margins_name="Total")
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([movie.director_name,movie.genres,],movie.country,rownames=['Genres','Director'],
            colnames=["Country"],
            dropna=True,margins=True)
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movie.genres,movie.color,values=movie.duration, aggfunc='mean',margins=True,margins_name="Total",dropna=True)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(movie.director_name,movie.color,values=movie.actor_likes, aggfunc='mean',margins=True,margins_name="Total",dropna=True)
```
